```python
import pandas as pd
import numpy as np

constructors = pd.read_csv('_final_project/1-Formula_One/constructors.csv')
results = pd.read_csv('_final_project/1-Formula_One/results.csv')

constructor_group = results.groupby(by = "constructorId")
constructor_mean_points = constructor_group["points"].mean()
display(constructor_mean_points)

mean_points_sorted = constructor_mean_points.sort_values(ascending = False)
display(mean_pts_sorted.iloc[:5])

constructors_subset = constructors[["constructorId", "name"]]

top5_constructors = pd.merge(mean_points_sorted, constructors_subset, on = "constructorId", how = "left")
display(top5_constructors.head(5))
#top 5 cars with best performance overall
```
