**Example:**

If we roll the following numbers (reading left-to-right):

$$ 1, 4, 3, 1, 1, 1, 2, 3, 2, 4, 3, 3$$

then `num_rolls` should have the value 12.

**Hint:** The computer needs to "remember" the previous dice roll. The most straightforward approach is to define a variable `previous_3 = False` and switch it to `True` when a 3 is rolled (don't forget to switch it back if you don't get two in a row!)

````python
num_rolls = 0

# your code here
import numpy as np
num_threes_row = 0
previous_3 = False

while num_threes_row < 2:
    if previous_3 == False:
        num_threes_row = 0
        roll = np.random.choice([1,2,3,4])
        num_rolls = num_rolls + 1
        if roll == 3:
            num_threes_row = num_threes_row + 1
            previous_3 == True
            continue
    roll = np.random.choice([1,2,3,4])
    num_rolls = num_rolls + 1
    if roll ==3:
        num_threes_row = num_threes_row + 1
    previous_3 == False
    continue
print("it took" ,num_rolls, "attempts to get two 3s in a row" )
    
````
