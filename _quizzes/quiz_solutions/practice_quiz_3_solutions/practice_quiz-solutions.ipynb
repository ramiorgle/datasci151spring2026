**Q2** Quadratic Formula

<font size = "3">

Consider solving the quadratic equation $ax^2 + bx + c = 0$. In addition to the standard quadratic formula, the following is an equivalent way of expressing the two solutions:

$$ x = \frac{2c}{-b \mp \sqrt{b^2 - 4ac}}$$

1. Write a function called `alt_quad_formula` that takes in 3 input arguments corresponding to the coefficients $a$, $b$, and $c$ and returns the two solutions to the corresponding quadratic equation computed using the expression above.

2. Test your function on the quadratic equation $3(x + 1)(x - 2) = 0$, verifying you get the outputs $-1$ and $2$.

````python
import numpy as np
### Your solution below

def alt_quad_formula(a, b, c):
    numer = 2*c 
    d_sqrt = np.sqrt(b**2 - 4*a*c)
    denom_1 = -b - d_sqrt 
    denom_2 = -b + d_sqrt 

    return numer/denom_1, numer/denom_2 

sol1, sol2 = alt_quad_formula(3, -3, -6)
print(sol1)
print(sol2)
````

**Q3** Function with no argument

<font size = "3">

Write a function that doesn't return any values (just like the `print` function).

The function takes in 3 arguments, with the following keywords:

- `vec_x`: A sequence of values (like a list, tuple, NumPy array, or column from a Pandas DataFrame).

- `xlabel`: A string

- `title`: A string

The function creates a histogram of the values in `vec_x`, labels the x-axis using the `xlabel` argument, always labels the y-axis with the string `"Frequency"`, and adds a title using the `title` argument. It then displays the histogram to the screen.

The function should have **no return values**. You can do this by having the last line of the function read `return None`. Or simply don't include a return statement at all.

````python
import matplotlib.pyplot as plt

def create_histogram(vec_x, xlabel, title):
    plt.hist(x = vec_x)
    plt.xlabel(xlabel)
    plt.ylabel("Frequency")
    plt.title(title)
    plt.show()
    return None # this is optional. Can make the previous line the final line of the function

````

**Q6** Split a dataset into subsets
 

<font size = "3">

- We will continue using the DataFrame you read in from "fifa23_players_basic.csv"

- Use ``len`` or ``.shape`` to define an integer variable called ``n``, which is the total number of rows in this DataFrame. <br>

- Use ``numpy.random.uniform`` to create a random sample of size ``n`` from the uniform distribution between 0 and 1. Then store these random numbers as a new column named "random_var" to the original DataFrame.<br> 
- Use ``.query`` to extract all players with ```random_var < 0.4```, and assign it to a new DataFrame called ``groupA``
- Use ``.query`` to extract all players with ```random_var >= 0.4```, and assign it to a new DataFrame called ``groupB``.

**Note**: In this way, we are assigning roughly 40% of the players to Group A, and roughly 60% to Group B

````python
import pandas as pd 
import numpy as np

df_fifa = pd.read_csv("quiz_data/fifa23_players_basic.csv")
n = len(df_fifa)
sample = np.random.uniform(low = 0.0, high = 1.0, size = n)
df_fifa["random_var"] = sample

groupA = df_fifa.query("random_var < 0.4")
groupB = df_fifa.query("random_var >= 0.4")

print(len(groupA))
print(len(groupB))


# You weren't asked to do the following, but it's a decent check on the assignment
percentA = len(groupA)/len(df_fifa) * 100
percentB = len(groupB)/len(df_fifa) * 100

print("Group A contains " + str(round(percentA, 2)) + "% of the data")
print("Group B contains " + str(round(percentB, 2)) + "% of the data")
````
