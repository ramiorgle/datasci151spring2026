```python
import pandas as pd
import numpy as np

constructors = pd.read_csv('_final_project/1-Formula_One/constructors.csv')
results = pd.read_csv('_final_project/1-Formula_One/results.csv')
races = pd.read_csv('_final_project/1-Formula_One/races.csv')

constructor_group = results.groupby(by = "constructorId")
constructor_mean_points = constructor_group["points"].mean()
display(constructor_mean_points)

mean_points_sorted = constructor_mean_points.sort_values(ascending = False)
display(mean_points_sorted.iloc[:5])

constructors_subset = constructors[["constructorId", "name"]]

top5_constructors = pd.merge(mean_points_sorted, constructors_subset, on = "constructorId", how = "left")
display(top5_constructors.head(5),)
#top 5 cars with best performance overall
```

```python
races_subset = races[["raceId", "year"]]
results_subset = results[["raceId", "constructorId","position"]]

results_year = pd.merge(results_subset, races_subset, on="raceId", how="left")
display(results_year)
#add code here to filter out the cars we are not analyzing.

plotting_data = results_year.query('constructorId == [131, 9, 23, 208, 6]')

display(plotting_data)
```

```python
# Rank top 5 drivers by mean points
drivers_subset = drivers[["driverId", "forename", "surname"]]

driver_mean_points = results.groupby(by="driverId")["points"].mean()
driver_mean_points_sorted = driver_mean_points.sort_values(ascending=False)

top5_drivers = pd.merge(driver_mean_points_sorted, drivers_subset, on="driverId", how="left")
display(top5_drivers.head(5))

# Find most frequent constructor for each top 5 driver
top5_driver_ids = top5_drivers["driverId"].iloc[:5].tolist()
top_driver_results = results[results["driverId"].isin(top5_driver_ids)]

constructor_counts = top_driver_results.groupby(["driverId", "constructorId"]).size()
most_frequent = constructor_counts.groupby(level="driverId").idxmax()
most_frequent = pd.DataFrame(most_frequent.tolist(), columns=["driverId", "constructorId"])

final = pd.merge(most_frequent, drivers_subset, on="driverId", how="left")
final = pd.merge(final, constructors_subset, on="constructorId", how="left")
display(final[["forename", "surname", "name"]])
```
