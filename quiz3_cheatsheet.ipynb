## Q1 — File looping with `os.listdir`

**What it does:** Gets all filenames in a folder, reads each one as a DataFrame, stores them in a list.

**Ctrl+F keywords:** `os.listdir`, `df_list`, `append`

**Common gotcha:** `os.listdir()` returns just the filename (e.g. `"germany_emission.csv"`), NOT the full path. You must manually join the folder path + filename.

```python
import os
import pandas as pd

# ✏️  Change this path to match the question
folder_path = "quiz_data/country_emissions"

# Step 1: get list of filenames in the folder
filenames = os.listdir(folder_path)

# Step 2: create empty list
df_list = []

# Step 3: loop over filenames
for filename in filenames:
    # IMPORTANT: join folder path + filename to get the full path
    full_path = folder_path + "/" + filename
    df = pd.read_csv(full_path)
    df_list.append(df)

# Step 4: after the loop
print(len(df_list))        # number of DataFrames
print(df_list[0].shape)    # shape of first DataFrame: (rows, cols)
```

## Q2 — Function that returns multiple values

**What it does:** Defines a math function with multiple inputs and multiple outputs.

**Ctrl+F keywords:** `def`, `return`, `np.sqrt`

**Common gotcha:** To use the two return values, unpack them: `x1, x2 = my_function(a, b, c)`

```python
import numpy as np

# Template: function with multiple inputs and multiple return values
def alt_quad_formula(a, b, c):
    # ✏️  Replace the body with whatever formula the question asks for
    x1 = (2 * c) / (-b - np.sqrt(b**2 - 4*a*c))
    x2 = (2 * c) / (-b + np.sqrt(b**2 - 4*a*c))
    return x1, x2   # return multiple values separated by a comma

# Unpack the two return values into two variables
# Test: 3(x+1)(x-2) = 0  →  expand to get  3x² - 3x - 6 = 0  →  a=3, b=-3, c=-6
sol1, sol2 = alt_quad_formula(3, -3, -6)
print(sol1)   # should be -1
print(sol2)   # should be  2
```

## Q3 — Function with NO return value (void function)

**What it does:** Defines a function that performs an action (like drawing a plot) but doesn't return anything — just like `print()`.

**Ctrl+F keywords:** `def`, `plt.hist`, `plt.show`, `plt.xlabel`

**Common gotcha:** Don't add a `return` statement. The function ends after `plt.show()`.

```python
import matplotlib.pyplot as plt

# Template: function that draws a histogram and returns nothing
# ✏️  The parameter names (vec_x, xlabel, title) must match what the question specifies
def create_histogram(vec_x, xlabel, title):
    plt.hist(vec_x)
    plt.xlabel(xlabel)        # x-axis label from the argument
    plt.ylabel("Frequency")   # y-axis label is always "Frequency"
    plt.title(title)          # title from the argument
    plt.show()
    # No return statement needed — the function returns None automatically

# How to call the function:
import pandas as pd
df = pd.read_csv("quiz_data/features.csv")

create_histogram(df["weight"],       "Weights",       "Weights in the Car Features dataset")
create_histogram(df["acceleration"], "Acceleration",  "Acceleration values in the Car Features dataset")
```

## Q4 — Subsetting a DataFrame with `.query()` and `.shape`

**What it does:** Filters rows based on a condition, counts how many pass the filter.

**Ctrl+F keywords:** `.query`, `.shape`

**Common gotchas:**
- `.shape` returns a *tuple* like `(18000, 12)`. Use `.shape[0]` to get just the number of rows as an integer.
- Column names with spaces need backticks inside the query string: `` `column name` ``
- To use a Python variable inside `.query()`, put `@` before the variable name.

```python
import pandas as pd

df = pd.read_csv("quiz_data/fifa23_players_basic.csv")

# ✏️  Change column name and threshold to match the question
df_tall = df.query("height_cm > 175")

# .shape returns (rows, cols) — use [0] to get just the row count as an integer
num_tall = df_tall.shape[0]
print(num_tall)   # should be an integer
print(type(num_tall))  # confirm: <class 'int'>

# --- Bonus: using a variable in .query() ---
threshold = 175
df_tall = df.query("height_cm > @threshold")  # @ before the variable name

# --- Bonus: multiple conditions ---
# df_subset = df.query("height_cm > 175 and Overall > 80")
# df_subset = df.query("height_cm > 175 or Overall > 90")
```

## Q5 — Define a function and apply it to a column with `.apply()`

**What it does:** Writes a classification function, then runs it on every row of a column.

**Ctrl+F keywords:** `.apply`, `def`, `if`

**Common gotcha:** Pass the function *name* to `.apply()` — no parentheses, no arguments: `.apply(classify_player)` NOT `.apply(classify_player())`

```python
import pandas as pd

df = pd.read_csv("quiz_data/fifa23_players_basic.csv")

# Step 1: define the classification function
# ✏️  Change the argument name, threshold, and return strings to match the question
def classify_player(overall):
    if overall > 85:
        return "Top Performer"
    else:
        return "Non Top-Performer"

# Step 2: extract the column you want to apply the function to
overall_col = df["Overall"]   # ✏️  Change column name to match the question

# Step 3: apply the function — pass the name only, NO parentheses
player_classification = overall_col.apply(classify_player)

print(player_classification)
```

## Q6 — Add a random column and split into groups

**What it does:** Generates random numbers, adds them as a new column, then uses `.query()` to split into two groups.

**Ctrl+F keywords:** `np.random.uniform`, `df["new_col"]`, `.query`

**Common gotcha:** The size of the random array must equal `n` (the number of rows). Get `n` from `.shape[0]` first.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("quiz_data/fifa23_players_basic.csv")

# Step 1: get total number of rows
n = df.shape[0]   # or: n = len(df)

# Step 2: generate n random numbers between 0 and 1, add as a new column
df["random_var"] = np.random.uniform(size=n)  # ✏️  Change column name if needed

# Step 3: split into groups using .query()
# ✏️  Change threshold (0.4) and column name to match the question
groupA = df.query("random_var < 0.4")   # ~40% of rows
groupB = df.query("random_var >= 0.4")  # ~60% of rows

print("Group A rows:", groupA.shape[0])
print("Group B rows:", groupB.shape[0])
print("Total:",        groupA.shape[0] + groupB.shape[0])  # should equal n
```
